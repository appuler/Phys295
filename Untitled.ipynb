#### The following is an experiment

The experiment attempts to model 2-dimensional Brownian Motion. First we will trying in a single plane (the x-y plane). Then, it will be followed by a three dimensional graph of the motion.

```python
import random
```

```python
x = [0]
y = [0]
x_sum = 0
y_sum = 0
for i in range(9):
    varx = random.randint(0, 1)
    vary = random.randint(0, 1)
    x_sum += varx*2 - 1
    y_sum += vary*2 - 1
    x.append(x_sum)
    y.append(y_sum)
```

```python
plt.plot(x, y, marker='o', linestyle='-')  # 'o' for markers, '-' for line

# Labels and title
plt.xlabel('X displacement')
plt.ylabel('Y displacement')
plt.title('Brownian Motion')
plt.grid(True)

plt.show()
```

```python
t = 1000
dd = 0.001

x = [0]
y = [0]
x_sum = 0
y_sum = 0
for i in range(t-1):
    varx = random.randint(0, 1)
    vary = random.randint(0, 1)
    x_sum += (varx*2 - 1) * dd
    y_sum += (vary*2 - 1) * dd
    x.append(x_sum)
    y.append(y_sum)
```

```python
plt.plot(x, y, linestyle='-')  # 'o' for markers, '-' for line

# Labels and title
plt.xlabel('X displacement')
plt.ylabel('Y displacement')
plt.title('Brownian Motion')
plt.grid(True)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
z = [i for i in range(1000)]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(z, x, y, c=z, cmap='coolwarm')

ax.set_xlabel('Time')
ax.set_ylabel('X Displacement')
ax.set_zlabel('Y Displacement')


plt.show()
```
