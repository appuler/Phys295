## About the Model

The Vasicek model is a mathematical model used to describe the evolution of interest rates over time. It is a type of short-rate model that models the instantaneous interest rate (short rate) as a stochastic process. The Vasicek model is widely used in finance for pricing bonds, interest rate derivatives, and managing risk.

Other Applications Include:
Bond Pricing: Used to determine the term structure of interest rates.
Risk Management: Helps in understanding the dynamics of interest rate changes.
Options & Derivatives: Applied in pricing interest rate derivatives like bond options and caps/floors.

The actual Model:
$$ dr = a(b - r)dt + \sigma dW$$

Where:

a = speed of mean reversion

b = long term mean

```python
def V_Model(sigma=0.1, a=0.1, b=0.041, R_initial=0.04, dt=0.001, total_time=1):
    R = [R_initial]
    W = [0]
    t_inc = int(total_time / dt)
    T = [dt*i for i in range(t_inc)]

    for i in range(t_inc - 1):
        dW = np.random.normal()*np.sqrt(dt)
        W.append(W[i] + dW)
        dR = a * (b - R[i])*dt + sigma*dW
        R.append(R[i] + dR)
        
    return R, W, T
```

```python
R1, W1, T1 = V_Model()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(T1, R1, color='#c19ce9')
plt.xlabel("Time")
plt.ylabel("Interest Rate")
plt.title(label="Vasicek Interest Rate Model Simulation 1")
plt.show()
```

```python
R2, W2, T2 = V_Model(sigma=0.02)
plt.figure(figsize=(10, 6))
plt.plot(T2, R2, color='#c19ce9')
plt.xlabel("Time")
plt.ylabel("Interest Rate")
plt.title(label="Vasicek Interest Rate Model Simulation 2")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(T2, W2, color='#c19ce9')
plt.xlabel("Time")
plt.ylabel("Brownian Motion")
plt.title(label="Vasicek Interest Rate Model Simulation 2")
plt.show()
```
