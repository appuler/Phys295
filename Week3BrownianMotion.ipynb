# Brownian Motion

The following text will be about my understanding of brownian motion and applications.

First we start with the idea of symmetric walks where we have a random Variable X_i such that at the ith fair coin flip, we have X_i = 1 when we land on H and X_i = -1 when we land on T. Then, we create a another random variable M which is the sum of X_i from 0 to some i where X_0 = 0. M represents the random walk on a smaller scale. It can be represented with the graph below.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import random
```

```python
x = [i for i in range(10)]
y = [0]
y_val = 0
for i in range(9):
    var = random.randint(0, 1)
    if var == 0:
        y_val -= 1
    if var == 1:
        y_val += 1
    y.append(y_val)
```

```python
plt.plot(x, y, marker='o', linestyle='-')  # 'o' for markers, '-' for line

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Connected Points')
plt.grid(True)

plt.show()
```

The graph above shows a simple random walk. However, to approximate Brownian Motion, we need to speed up time and scale down the step size. Here the step size is 1 which is how much it moves at each time. 

#### Scaled Symmetric Walks

The actual function becomes W^(n)(t) = 1/sqrt(n) M_(nt) where n approaches infinity and t approaches 0. The graph below will demonstrate an approximation of brownian motion by decreasing the step size to 0.001 and increasing the number of times the flip is played to 1000.

```python
step = 0.001
t = 1000

x = [i for i in range(t)]
y = [0]
y_val = 0
for i in range(t - 1):
    var = random.randint(0, 1)
    if var == 0:
        y_val -= step
    if var == 1:
        y_val += step
    y.append(y_val)
```

```python
plt.plot(x, y, linestyle='-')
# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Connected Points')
plt.grid(True)

plt.show()
```

#### Additional Notes

The probability distribution of each point of time is a binomial distribution. That means that as the graph's time scale approaches infinity, the probability distribution converges to a normal distribution by the central limit theorem. 

Additionally, this is simply an approximation of what Brownian motion is but is unable to reach true brownian motion. 

Properties of brownian motion include:
- Being finite
- Being Continuous
- Satisfying both the Markov and Martingale Properties
- Having unbounded variation
- Being strongly normally distributed

The properties that our random walk breaks is being continuous.
